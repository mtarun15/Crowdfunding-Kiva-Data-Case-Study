## Project Background

The Crowdfunding Phenomenon The process of raising money to support a project or business venture through a large number of investors and via an online platform is known as crowdfunding. With the advent of online crowdfunding, there are now more avenues for individuals, businesses, and organizations to access capital. In theory, platforms for crowdfunding are created to connect people who are willing to lend or invest their money with other people, projects, or businesses who need financial support.

#### Process flow diagram for funding:

##### -Reference- IESE Business School-University of Navarra

Kiva.org One of the biggest online crowdfunding sites, Kiva.org was established in 2005 and allows anyone to lend money to untapped entrepreneurs all over the world. Kiva's aim is to "connect people via lending to relieve poverty." In areas where traditional banking services are unavailable, Kiva works to give secure, inexpensive access to money so that individuals can start their own businesses and better their lives and those of their families. Kiva basically serves as a conduit online that connects lenders and borrowers. On Kiva's platform, profiles of people from underdeveloped nations who are looking for microcredit are posted. The qualities of the loan request and the borrower are taken into consideration when lenders peruse the various profiles and make investments in the projects they find most appealing. The borrower's gender, country of origin, and other loan features, such as the loan's intended use and payback duration, may all influence how much money they receive based on how well it fits with the lender's preferences. Neither Kiva nor the lenders it connects to the borrowers get interest from the loans they arrange. However, borrowers pay interest to the "field partners," who are effectively the ones that face the majority of the risk in the financing process. When compared to many other loan systems, Kiva stands out for its work with field partners. The field partners' job is to streamline the process by creating a local presence and disbursing funds to the borrowers before the loan request is posted on the Kiva platform, giving them a head start on their business endeavor.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,target, test_size=.33,random_state = 1)
```

Attempting to fit a logistic regression model.

```python
from sklearn.linear_model import LogisticRegression
logregressor=LogisticRegression(random_state=1)
logregressor.fit(X_train,Y_train)
```

Following along, we got below results for the test dataset.

```python
y_pred=logregressor.predict(X_test)
y_pred[:10]
```

We used Scikit learning to evaluate the model using accuracy rate and confusion matrix.

```python
from sklearn.metrics import accuracy_score
accuracy_score ( y_true=Y_test,y_pred=y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true=Y_test,y_pred=y_pred)
```
